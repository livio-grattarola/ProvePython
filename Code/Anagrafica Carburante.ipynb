# DimProvince

```python
#Lettura csv Regioni e Province
import pandas as pd
province = pd.read_csv("dati_carburante/Province.csv", sep=";", keep_default_na=False, na_values=['']);
province.columns = ["Prov_" + x for x in province.columns]

regioni =  pd.read_csv("dati_carburante/Regioni.csv", sep=";");
regioni.columns = ["Reg_" + x for x in regioni.columns] 

```

```python
#Creazione csv DimProvince
import numpy as np
fullProvinces = province.merge(regioni, left_on=['Prov_CodiceIstatRegione'], right_on=['Reg_CodiceIstat'], how='left')
del fullProvinces['Prov_CodiceIstatCM']
del fullProvinces['Prov_CM']
del fullProvinces['Prov_CodiceIstatRegione']
del fullProvinces['Reg_Costiera']
fullProvinces = fullProvinces.rename({'Reg_Abbreviato': 'Reg_Short'})

fullProvinces.to_csv("./out/DimProvince.csv", sep=';', index=False)
fullProvinces
```
